## Implement matrix-vector multiplication from scratch

```python
def mat_vec_mult(mat, vec):
    if mat.shape[1] != vec.shape[0]:
        raise ValueError(f"Dim mismatch: {mat.shape[1]} != {vec.shape[0]}")
    
    res = np.zeros(mat.shape[0], dtype=mat.dtype)
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            res[i] += mat[i, j] * vec[j] 
    return res
```

```python
print("\n[mat_vec_mult] Test")
try:
    mat1 = np.array([[1, 2], [3, 4]])
    vec1 = np.array([1, 0])
    res1 = mat_vec_mult(mat1, vec1)
    assert np.allclose(res1, np.array([1., 3.])), f"Expected [1., 3.], got {res1}"
    
    mat2 = np.array([[2, -1, 3], [0, 5, -2]])
    vec2 = np.array([1, 1, 1])
    res2 = mat_vec_mult(mat2, vec2)
    assert np.allclose(res2, np.array([4., 3.])), f"Expected [4., 3.], got {res2}"
    print("✅ mat_vec_mult passed!")
except Exception as e:
    print(f"❌ mat_vec_mult failed: {e}")
```
